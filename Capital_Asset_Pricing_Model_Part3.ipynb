### Excess stock and market returns

```python
# go through each stock and calculate the excess return
for share in shares:
  historical_data[share + '_ER'] = daily_return[share] - rfr
historical_data.dropna(inplace=True)
print(historical_data.head())
```

### Running CAPM

$$ E(R_{share}) =R_{f} + \beta_{share}(E(R_{mkt}-R_{f})) $$  $$ E(R_{share}) - R_{f} = \alpha + \beta_{share}(E(R_{mkt}-R_{f}))  $$



*   Risk premium of individual stock depends on 
 * market risk 
 * market risk-premium



```python
from scipy import stats
d = []
# compute linear regression of the stocks
for s in shares:
  (beta, alpha, r_value, p_value, std_err) = stats.linregress(historical_data['^GSPC_ER'],historical_data["{0}_ER".format(s)])
  d.append(
      {
       "share" : s,  
       "beta": beta,
       "alpha": alpha,
       "r_value"   : r_value,
       "p_value"  : p_value,
       "std_err"  : std_err
      }
  )

plt.rcParams["figure.figsize"] = (20,5)
fig, ax = plt.subplots()

# hide axes
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
df = pd.DataFrame(d)
ax.table(cellText=df.values, colLabels=df.columns, loc='center')
fig.tight_layout()
plt.show()
```
